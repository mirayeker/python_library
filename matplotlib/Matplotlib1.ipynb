```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline 
```

```python
yasListesi=[10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi=[20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi=np.array(yasListesi)
numpyKiloListesi=np.array(kiloListesi)
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"b")
plt.xlabel("Yas")
plt.ylabel("Kilo")
#plt.show() # baska bi editörde yazsaydık bunu yazmamız lazımdı görebilmek için
plt.title("Kilonun yasa gore degisim grafigi")
```

```python
numpyDizisi1=np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2=numpyDizisi1 **3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"b*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"g*-")
plt.subplot(1,2,2)# 1 sıra olacak, 2 kolon olcak, su an 2 grafiği ciziyorum
plt.plot(numpyDizisi2,numpyDizisi1,"r--")
```

```python
benimFigure=plt.figure()
figureAxes=benimFigure.add_axes([0.1,0.1,0.3,0.3])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X ekseni ")
figureAxes.set_ylabel("Y ekseni")
figureAxes.set_title("Grafik basligi")
```

```python
figure2=plt.figure()
eksen1=figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2=figure2.add_axes([0.2,0.4,0.3,0.3])
eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("Xekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana Grafik basliği")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("Xekseni")
eksen2.set_ylabel("Y ekseni")
eksen2.set_title("kucuk Grafik basliği")
```

```python
(benimFigure, benimEksenler)=plt.subplots(nrows=1,ncols=2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"b")
    eksen.set_xlabel("X ekseni")
    
plt.tight_layout()
```

```python
yeniFigur=plt.figure(dpi=150)
#dpi=150 
#yeniFigur.savefig("benimfigur",dpi=200)

yeniEksen=yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 **2,label="numpyDizisi **2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 **3,label="numpyDizisi **3")
yeniEksen.legend(loc=6)
```
